1 (i) KNN Regression:

Perform k-nearest neighbor regression for this dataset using both normalized
and raw features. Find the value of k ∈ {1, 2, . . . , 100} that gives you the
best fit. Plot the train and test errors in terms of 1/k.


```python
testing_error_rawdata = []
training_error_rawdata = []

for k in range(1,101,1):
    knn_rawdata = KNeighborsRegressor(n_neighbors=k)
    knn_rawdata.fit(x_train,y_train)
    
    knn_rawdata_predict_test = knn_rawdata.predict(x_test)
    test_cm_rawdata = mean_squared_error(y_test,knn_rawdata_predict_test)
    testing_error_rawdata.append(test_cm_rawdata)

    knn_rawdata_predict_train = knn_rawdata.predict(x_train)
    train_cm_rawdata = mean_squared_error(y_train,knn_rawdata_predict_train)
    training_error_rawdata.append(train_cm_rawdata)

# print(testing_error_rawdata)
# print(training_error_rawdata)  
# print(min(testing_error_rawdata))
# print(min(training_error_rawdata))
```

```python
print("Best k for testing raw data")
print("k = " , testing_error_rawdata.index(min(testing_error_rawdata))+1)
print("Test Error = " , min(testing_error_rawdata))

print("Best k for training raw data")
print("k = " ,  training_error_rawdata.index(min(training_error_rawdata[1:]))+1)
print("Train error = " , min(training_error_rawdata[1:]))
```

```python
k = np.linspace(1,100,100)
plt.plot(1/k, testing_error_rawdata)
plt.plot(1/k, training_error_rawdata)
plt.xlim(max(1/k), min(1/k))
plt.xlabel("1/k_values")
plt.ylabel("error")
plt.legend(["Test error", "Train error"])
```

## HW2(2) 2.4.1

For each of parts (a) through (d), indicate whether we would generally expect the performance of a flexible statistical learning method to be better or worse than an inflexible method. Justify your answer.

(a) The sample size n is extremely large, and the number of predictors p is small.

(b) The number of predictors p is extremely large, and the number of observations n is small.

(c) The relationship between the predictors and response is highly non-linear.

(d) The variance of the error terms, i.e. σ2 = Var(ϵ), is extremely high.

Answers :

(a) Better - The flexible statistical learning method will perform "better" as there would be more samples to understand the predictors and also reduces overfiiting.

(b) Worse - With small number of observations "n", likely to overfit the data. The inflexible model would perform better.

(c) Better -  Flexible method perfoms better with non-linear data. 

(d) Worse - The high variance indicate that lot of noise or error, so inflexible model perform better. A flexible model might overfit.

## HW2(2) 2.4.7

The table below provides a training data set containing six observations, three predictors, and one qualitative response variable.

                                              Obs. X1 X2 X3  Y
                                              1     0  3  0 Red
                                              2     2  0  0 Red
                                              3     0  1  3 Red
                                              4     0  1  2 Green
                                              5    −1  0  1 Green
                                              6     1  1  1 Red

Suppose we wish to use this data set to make a prediction for Y when X1 = X2 = X3 = 0 using K-nearest neighbors.

(a) Compute the Euclidean distance between each observation and the test point, X1 = X2 = X3 = 0.

(b) What is our prediction with K = 1? Why?

(c) What is our prediction with K = 3? Why?

(d) If the Bayes decision boundary in this problem is highly non-linear, then would we expect the best value for K to be large or small? Why?

Answers :

(a) The Euclidean distacne of a 3D space can be found out using the following formula:<br>
                $\sqrt(x1 - \bar{x1})^2 + (x2 - \bar{x2})^2 + (x3 - \bar{x3})^2$

Obs 1 : $\sqrt {(0-0)^2 + (3-0)^2 + (0-0)^2}  = 9 $ <br>
Obs 2: $\sqrt {(2-0)^2 + (3-0)^2 + (0-0)^2}  = 2 $ <br>
Obs 3: $\sqrt {(0-0)^2 + (1-0)^2 + (3-0)^2}  = \sqrt{10} = 3.16 $ <br>
Obs 4: $\sqrt {(0-0)^2 + (1-0)^2 + (2-0)^2}  = \sqrt{5} = 2.24 $ <br>
Obs 5: $\sqrt {(-1-0)^2 + (0-0)^2 + (1-0)^2}  = \sqrt{2} = 1.41 $ <br>
Obs 6: $\sqrt {(1-0)^2 + (1-0)^2 + (1-0)^2}  = \sqrt{3} = 1.73 $ <br>

(b) Green : With k = 1, the nearest neighbour is Obs.5 (-1,0,1) Also, the Euclidean distacne calculated for Obs.5 is the lowest one.

(c) Red : With k = 3, the nearest neighbours are Obs.5(−1,0,1) , Obs.6(1,1,1) and Obs.2(2,0,0), with the least Euclidean distacne from the test point.

(d) A higher non-linear Bayes boundary expects a small value of k. So, smaller k values are better.
