iii. Break each time series in your training set into l ∈ {1, 2, . . . , 20} time series
of approximately equal length and use logistic regression5
to solve the binary
classification problem, using time-domain features. Remember that breaking
each of the time series does not change the number of instances. It only
changes the number of features for each instance. Calculate the p-values for
your logistic regression parameters in each model corresponding to each value
of l and refit a logistic regression model using your pruned set of features.6
Alternatively, you can use backward selection using sklearn.feature selection
or glm in R. Use 5-fold cross-validation to determine the best value of the pair
(l, p), where p is the number of features used in recursive feature elimination.
Explain what the right way and the wrong way are to perform cross-validation
in this problem.7 Obviously, use the right way! Also, you may encounter the
problem of class imbalance, which may make some of your folds not having
any instances of the rare class. In such a case, you can use stratified cross

```python
#The following code has been used from the below URL. 
#This code helps in spliting the dataframe(df) and splits is the number of splits to the dataframe 
# Source: https://datagy.io/split-pandas-dataframe/ 
def split_dataframe_by_position(df, splits):
    """
    Takes a dataframe and an integer of the number of splits to create.
    Returns a list of dataframes.
    """
    dataframes = []
    index_to_split = len(df) // splits
    start = 0
    end = index_to_split
    for split in range(splits):
        temporary_df = df.iloc[start:end, :]
        dataframes.append(temporary_df)
        start += index_to_split
        end += index_to_split 
    return dataframes
```

Train set - 2aiii

```python
#This code is for training data set 
#Here the data frame is split l= 1..20, extract features, implement logistic regression and save the values
trainingframe_list = [train_bending1, train_bending2,train_cycling, train_lying, train_sitting, train_standing, train_walking]
lp_vales = {}
list_of_timeseries_df = []
AUC_list = []
best_feature = []
acc_auc_dict = {}
fitted_models = []

#Iterating for l splits
for l in range(1,21):
    timeseries_df = pd.DataFrame()
    length_df = 0
    #Get the train data from the list
    for traindataset in trainingframe_list:
        #Get each file 
        for file in traindataset:
            
            split = split_dataframe_by_position(file,l)
            
            for i in range(len(split)):
                split[i].reset_index(drop = True, inplace=True)
            
            
            #Combine the splits
            df_combined = pd.concat(split,axis = 1, ignore_index= True)
            
            #Feature extraction
            for eachcol in df_combined:
                timeseries_df.loc[length_df, "min"+str(eachcol)] = df_combined[eachcol].min()
                timeseries_df.loc[length_df, "max"+str(eachcol)] = df_combined[eachcol].max()
                timeseries_df.loc[length_df, "std dev"+str(eachcol)] = df_combined[eachcol].std()
            length_df += 1
        list_of_timeseries_df.append(timeseries_df)
    for i in range(len(timeseries_df)):
        if i < 9:
            timeseries_df.loc[i,"class"] = 1
        else:
            timeseries_df.loc[i,"class"] = 0
            
    x = timeseries_df.iloc[:,:-1]
    y = timeseries_df.iloc[:,-1:].values.ravel()
    
    #Logistic regression with 5 fold cross validation
    
    estimator = LogisticRegression( max_iter = 100000 , C = sys.maxsize)
    selector = RFECV(estimator, cv=StratifiedKFold(5))
    selector_fit = selector.fit(x,y)
    
    fitted_models.append(selector_fit)
    
    ranking = selector_fit.ranking_
    
    selected_features = []
    #select the best feature according to ranking
    for index,rank in enumerate(ranking):
        if rank == 1:
            selected_features.append(timeseries_df.columns[index])

    best_feature.append(selected_features)
    
    new_x = x.loc[:,selected_features]
    
    new_model = LogisticRegression(solver="liblinear",C = sys.maxsize)
    new_model.fit(new_x,y)
    
    #Compute area under the curve
    accuracy = np.mean(selector_fit.cv_results_["mean_test_score"])
    AUC = roc_auc_score(y,new_model.predict_proba(new_x)[:,1])
    
    acc_auc_dict[l] = {
        "Cross Validation Accuracy" : accuracy,
        "AUC" : AUC
    }
        
    lp_vales[str(l)] = {
        "L value" : l,
        "P value" : len(selected_features)
    }

```

```python
#Claculate the max accuracy, the split position and the best features used by backward elimination of features
acc_auc_df = pd.DataFrame(acc_auc_dict).T
acc_auc_df.index.name = "L"
max_accuracy = 0
max_auc = 0
split = 0
for i,auc in enumerate(acc_auc_df["AUC"]):
        if max_auc < auc:
            max_auc = auc
            best_feature_for_l = best_feature[i]
            split = i
            max_accuracy = acc_auc_df.iat[i,0]

print(f"Best Feature(s): {best_feature_for_l}")
print(f"Number of Splits: {split + 1}")

acc_auc_df
```

```python
#The following is used to calculate the confusion matrix and the ROC curve with the AUC score.
#The β values and the p -values are also displayed below which is extracted using the coef function
x = list_of_timeseries_df[split].iloc[:,:-1]
y = list_of_timeseries_df[split].iloc[:,-1:]
y_predict = fitted_models[split].predict(x)

print("Confusion Matrix :")
print(confusion_matrix(y,y_predict))

#Calculate p values
_ , pValues = chi2(x,y)

auc_score = roc_auc_score(y, y_predict)
print(f"AUC score :  {auc_score}")

plot_roc_curve(fitted_models[split],x,y)
plt.title("AUC ROC curve")
plt.show()

model = fitted_models[split]
for parameter in model.estimator_.coef_:
    for index,val in enumerate(parameter):
        print(f"β{index} : {val} -- pValue : {pValues[index]}")
```

```python
trainingframe_list = [train_bending1, train_bending2,train_cycling, train_lying, train_sitting, train_standing, train_walking]
multiclass_names = ["bending","bending","cycling","lying","sitting","standing","walking"]
list_of_timeseries_df = []
list_normalized_X = []
list_y = []
list_fitted_models_normalized_Multinomial = []

acc_auc_dict = {}
fitted_models_l1 = []

#Iterating for l splits
for l in range(1,21):
    timeseries_df = pd.DataFrame()
    length_df = 0
    #Get the train data from the list
    for class_name,traindataset in enumerate(trainingframe_list):
        #Get each file 
        for file in traindataset:
            
            split = split_dataframe_by_position(file,l)
            
            for i in range(len(split)):
                split[i].reset_index(drop = True, inplace=True)
            
            
            #Combine the splits
            df_combined = pd.concat(split,axis = 1, ignore_index= True)
            
            #Feature extraction
            for eachcol in df_combined:
                timeseries_df.loc[length_df, "min"+str(eachcol)] = df_combined[eachcol].mean()
                timeseries_df.loc[length_df, "max"+str(eachcol)] = df_combined[eachcol].median()
                timeseries_df.loc[length_df, "std dev"+str(eachcol)] = df_combined[eachcol].std()
            
            timeseries_df.loc[length_df,"Multi_class"] = multiclass_names[class_name]
            length_df += 1
        list_of_timeseries_df.append(timeseries_df)

    x = timeseries_df.iloc[:,:-1]
    y = timeseries_df.iloc[:,-1:].values.ravel()

    #Normalize the data
    scaler = MinMaxScaler()
    normalizedX = scaler.fit_transform(x,y)
    
    list_normalized_X.append(normalizedX)
    list_y.append(y)
    
    #Naive bayes classifier 
    Multinomial_model = MultinomialNB().fit(normalizedX,y)
    list_fitted_models_normalized_Multinomial.append(Multinomial_model)
    #Compute area under the curve
    accuracy = Multinomial_model.score(normalizedX,y)
    
    acc_auc_dict[l] = {
        "Multinomial Accuracy Multiclass" : accuracy
    }
```
