The different types of time domain features usually used in time series classification are :

Simple statistical features:
* Mean for the dimensions
* Standard deviations for the dimensions
* Skewness, Kurtosis
* Maximum and minimum

Time series analysis related features:
* Cross Correlations and Auto Correlations
* Orders of auto regression, integrated and moving average of the model
* Parameters of auto regression
* Parameters of moving avarage

Frequence domain realted features:
* Frequencies for k peaks
* K-quantiles

c(iii) Estimate the standard deviation of each of the time-domain features you extracted from the data. Then, use Python’s bootstrapped or any other method to build a 90% bootsrap confidence interval for the standard deviation of each feature.

```python
#Here we are using python bootstrap to obtain the Confidence low and high values and the stanard Error 
# The bootstrap(std) method is used on the previously obatined DataFrame timedomainfeatures_df
#All these values are stored in the bootstrap_CI_df. The confidence level is at 90%
std_bootstrap = {}
for column_features in timedomainfeatures_df.columns:
    bootstrap_confidance = bootstrap((timedomainfeatures_df[column_features],), 
                                     np.std, 
                                     confidence_level=0.9, 
                                     n_resamples=1000, 
                                     method="percentile"
                                    )
    std_bootstrap[str(column_features)] = {"Confidence Low" : bootstrap_confidance.confidence_interval.low,
                                           "Confidence High" : bootstrap_confidance.confidence_interval.high,
                                           "Standard Error" : bootstrap_confidance.standard_error}
    
bootstrap_CI_df = pd.DataFrame(std_bootstrap).T
bootstrap_CI_df
```
